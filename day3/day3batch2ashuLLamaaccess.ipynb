```python
# import required modules
import json
from transformers import AutoTokenizer, AutoModelForCausalLM,pipeline
```

```python
# reading token from json file
mydata = json.load(open("mytoken.json"))
hg_token = mydata["HF_TOKEN"]
```

```python
# select the model
model_name = "meta-llama/Llama-3.2-1B"
```

```python
# use our token to interface with model using transformer by hugging face
tokenizer = AutoTokenizer.from_pretrained(model_name,
                                          use_auth_token=hg_token)
```

```python
# we need to setup device to LLama model which gonna used like CPU , GPU using automodelforcausual
model = AutoModelForCausalLM.from_pretrained(model_name,
                                             use_auth_token=hg_token,
                                             device_map="auto" # is set what hardware like CPU,GPU
                                             )
```

```python
# set model mode to eval since we want to do query
model = model.eval()
```

```python
# setting text-generation pipeline in model
pipe = pipeline(
    "text-generation",
     model=model,
     tokenizer=tokenizer,
     max_new_tokens=512
  )
```

```python
# response funcation using pipe
def get_response(prompt):
  response1 = pipe(prompt)
  return response1[0]["generated_text"]
```

```python
# lets query get_response function
my_query = "explain is machine learning in 100 words"
llama3_response = get_response(my_query)
print(llama3_response)
```
